Importing libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading data:

```python
df = pd.read_csv("C:/Users/Pritam/Desktop/Project 4th sem/Data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().mean()*100
```

Checking if CUST_ID contains duplicates or not:

```python
df["CUST_ID"].nunique()
```

Dropping CUST_ID column:

```python
df.drop("CUST_ID", axis=1, inplace=True)
```

Dropping the NA values in CREDIT_LIMIT column:

```python
df.dropna(subset=["CREDIT_LIMIT"], inplace=True)
```

Imputing the NA values in MINIMUM_PAYMENTS column by its median:

```python
df["MINIMUM_PAYMENTS"].fillna(df["MINIMUM_PAYMENTS"].median(), inplace=True)
```

```python
plt.figure(figsize=(20,35))
for i, col in enumerate(df.columns):
    if df[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sns.kdeplot(df[col], ax=ax)
        plt.xlabel(col)
        
plt.show()
```

Handling skewness:

```python
cols = ['BALANCE','PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'ONEOFF_PURCHASES_FREQUENCY','PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT']
for col in cols:
    df[col] = np.log(1 + df[col])
```

```python
plt.figure(figsize=(15,25))
for i, col in enumerate(cols):
    ax = plt.subplot(7, 2, i+1)
    sns.kdeplot(df[col], ax=ax)
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
